# Data Preprocessing / Feature Engineering

```python
def clean_review(review): 
    
    # changing to lower case
    lower = review.str.lower()
    
    # Replacing the repeating pattern of &#039;
    pattern_remove = lower.str.replace("&#039;", "")
    
    # Removing all the special Characters
    special_remove = pattern_remove.str.replace(r'[^\w\d\s]',' ')
    
    # Removing all the non ASCII characters
    ascii_remove = special_remove.str.replace(r'[^\x00-\x7F]+',' ')
    
    # Removing the leading and trailing Whitespaces
    whitespace_remove = ascii_remove.str.replace(r'^\s+|\s+?$','')
    
    # Replacing multiple Spaces with Single Space
    multiw_remove = whitespace_remove.str.replace(r'\s+',' ')
    
    # Replacing Two or more dots with one
    dataframe = multiw_remove.str.replace(r'\.{2,}', ' ')
    
    return dataframe
```

```python
df['clean_review'] = clean_review(df['review'])
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# Removing the stopwords
stop_words = set(stopwords.words('english'))
df['clean_review'] = df['clean_review'].apply(lambda x: ' '.join(word for word in x.split() if word not in stop_words))
```

```python
# Removing the word stems using the Snowball Stemmer
from nltk.stem import SnowballStemmer
Snow_ball = SnowballStemmer("english")
df['clean_review'] = df['clean_review'].apply(lambda x: " ".join(Snow_ball.stem(word) for word in x.split()))
```

# Converting text to matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(df['clean_review'])
```

```python
Y = df['sentiment']
```

# Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y ,test_size=0.3, random_state=42)
```

```python
print ("The Train set size ", X_train.shape)
print ("The Test set size ", X_test.shape)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Initialize XGBoost Classifier
xgb = XGBClassifier(
    n_estimators=1000,  # This parameter specifies the number of boosting rounds (trees) to build
    learning_rate=0.1,  # it controls the step size shrinkage used to prevent overfitting
    max_depth=5,        #  It determines the maximum depth of the individual trees
    reg_alpha=0.01,     # L1 regularization term on weights
    reg_lambda=0.01,    # L2 regularization term on weights.
    min_child_weight=1, 
    subsample=0.8,      #  Subsampling can speed up training and reduce overfitting by training on a fraction of the data.
    colsample_bytree=0.8,
    n_jobs=-1)          # Setting it to -1 uses all available cores, speeding up training.

```

```python
# Fit the model
xgb.fit(X_train, Y_train)
```
